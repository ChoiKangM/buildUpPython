```python
%matplotlib inline
import matplotlib.pyplot as plt

x = range(100)
y = [i*i for i in x]

plt.plot(x, y, 'g+')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```

```python
import math

x = range(5)
sin_y = [math.sin(2*3.14*i) for i in x]
cos_y = [math.cos(2*3.14*i) for i in x]

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')

plt.show()
```

```python
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

t = np.arange(0.0, 5.0, 0.1)
ax1.plot(t, np.sin(2*np.pi*t), 'b--')
ax2.plot(t, np.cos(2*np.pi*t), 'r--')

plt.show()
```
