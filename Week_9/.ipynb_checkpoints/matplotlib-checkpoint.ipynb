```python
%matplotlib inline
import matplotlib.pyplot as plt

x = range(100)
y = [i*i for i in x]

plt.plot(x, y, 'g+')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```
